```python
# Import the dataset
import pandas as pd
from google.colab import files
file = files.upload()  #upload file into google colab session
df = pd.read_csv("credit-cards.csv") 
df.head()
```

```python
from google.colab import drive
drive.mount('drive')
df['SEX'] = df['SEX'].replace([1,2], ['male','female'])
df['EDUCATION'] = df['EDUCATION'].replace([0,1,2,3,4,5,6], ['Others','Master','Bachelor','High School','Others','Others','Others'])
df['MARRIAGE'] = df['MARRIAGE'].replace([0,1,2,3], ['others','married','single','divorce'])
df.to_csv('credit.csv')
!cp credit.csv "drive/My Drive/"
```

```python
df['PAY_0_new'] = df['PAY_0'].replace([1,2,3,4,5,6,7,8], [1,1,1,1,1,1,1,1])
df['PAY_0_new'] = df['PAY_0_new'].replace([-2,-1,0], [-1,-1,-1])

df['PAY_2_new'] = df['PAY_2'].replace([1,2,3,4,5,6,7,8], [1,1,1,1,1,1,1,1])
df['PAY_2_new'] = df['PAY_2_new'].replace([-2,-1,0], [-1,-1,-1])

df['PAY_3_new'] = df['PAY_3'].replace([1,2,3,4,5,6,7,8], [1,1,1,1,1,1,1,1])
df['PAY_3_new'] = df['PAY_3_new'].replace([-2,-1,0], [-1,-1,-1])

df['PAY_4_new'] = df['PAY_4'].replace([1,2,3,4,5,6,7,8], [1,1,1,1,1,1,1,1])
df['PAY_4_new'] = df['PAY_4_new'].replace([-2,-1,0], [-1,-1,-1])

df['PAY_5_new'] = df['PAY_5'].replace([1,2,3,4,5,6,7,8], [1,1,1,1,1,1,1,1])
df['PAY_5_new'] = df['PAY_5_new'].replace([-2,-1,0], [-1,-1,-1])

df['PAY_6_new'] = df['PAY_6'].replace([1,2,3,4,5,6,7,8], [1,1,1,1,1,1,1,1])
df['PAY_6_new'] = df['PAY_6_new'].replace([-2,-1,0], [-1,-1,-1])
df
```

```python
###PAY_0
df1 = df.groupby(['EDUCATION','PAY_0_new']).count().reset_index()
df1 = df1[['EDUCATION','PAY_0_new','ID']]
df1.columns = ['EDUCATION','PAY_duly','Number']
df1
```

```python
###PAY_2
df2 = df.groupby(['EDUCATION','PAY_2_new']).count().reset_index()
df2 = df2[['EDUCATION','PAY_2_new','ID']]
df2.columns = ['EDUCATION','PAY_duly','Number']
df2
```

```python
###PAY_3
df3 = df.groupby(['EDUCATION','PAY_3_new']).count().reset_index()
df3 = df3[['EDUCATION','PAY_3_new','ID']]
df3.columns = ['EDUCATION','PAY_duly','Number']
df3
```

```python
###PAY_4
df4 = df.groupby(['EDUCATION','PAY_4_new']).count().reset_index()
df4 = df4[['EDUCATION','PAY_4_new','ID']]
df4.columns = ['EDUCATION','PAY_duly','Number']
df4
```

```python
###PAY_5
df5 = df.groupby(['EDUCATION','PAY_5_new']).count().reset_index()
df5 = df5[['EDUCATION','PAY_5_new','ID']]
df5.columns = ['EDUCATION','PAY_duly','Number']
df5
```

```python
###PAY_6
df6 = df.groupby(['EDUCATION','PAY_6_new']).count().reset_index()
df6 = df6[['EDUCATION','PAY_6_new','ID']]
df6.columns = ['EDUCATION','PAY_duly','Number']
df6
```

```python
###PAY_0,PAY_2,PAY_3,PAY_4,PAY_5,PAY_6
from google.colab import drive
drive.mount('drive')
result = pd.merge(df1, df2, on=['EDUCATION','PAY_duly'])
result = pd.merge(result, df3, on=['EDUCATION','PAY_duly'])
result = pd.merge(result, df4, on=['EDUCATION','PAY_duly'])
result = pd.merge(result, df5, on=['EDUCATION','PAY_duly'])
result = pd.merge(result, df6, on=['EDUCATION','PAY_duly'])
result.columns = ['EDUCATION','PAY_duly','Number0','Number2','Number3','Number4','Number5','Number6']
result.to_csv('pay.csv')
!cp pay.csv "drive/My Drive/"
```
